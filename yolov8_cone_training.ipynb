```python
from ultralytics import YOLO
import os
from sklearn.model_selection import train_test_split
import shutil

# Define paths
images_dir = '/content/drive/MyDrive/dataset/images/train'
labels_dir = '/content/drive/MyDrive/dataset/labels/train'

val_images_dir = '/content/drive/MyDrive/dataset/images/val'
val_labels_dir = '/content/drive/MyDrive/dataset/labels/val'

# Get list of image files
image_files = [f for f in os.listdir(images_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Split images into train and validation sets
train_imgs, val_imgs = train_test_split(image_files, test_size=0.2, random_state=42)

print(f"Total images: {len(image_files)}")
print(f"Training images: {len(train_imgs)}")
print(f"Validation images: {len(val_imgs)}")

# Move validation images and labels
for img_file in val_imgs:
    # Move image
    shutil.move(os.path.join(images_dir, img_file), os.path.join(val_images_dir, img_file))

    # Move corresponding label file
    label_file = img_file.replace('.png', '.txt').replace('.jpg', '.txt').replace('.jpeg', '.txt')
    if os.path.exists(os.path.join(labels_dir, label_file)):
        shutil.move(os.path.join(labels_dir, label_file), os.path.join(val_labels_dir, label_file))
    else:
        print(f"Warning: Label file {label_file} not found for image {img_file}")

model = YOLO("yolov8n.pt")

model.train(
    data="/content/drive/MyDrive/dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=2
)
```
